### Create a `RollDie` Python Class


Usage Ex.  `mydie = RollDie(4)`   # create a 4 sided die with possible values 1,2,3,4, and randomly set starting value

The class should provide the following methods for an instance `mydie`:
 - `mydie.size()`: return the number of sides of mydie
 - `mydie.get_value()`:  return the current value of mydie
 - `mydie.set_value(2)`:  set the current value of mydie to 2
 - `mydie.roll()`:  randomly set a new value of mydie and return that value

```python
import numpy as np
class RollDie:
    """A class to roll and set a die of any number of sides.
    
    Usage:  Initialize an object, pass nunber of sides of die
    
    Ex.  mydie = RollDie(4)   # create an 4 sided die with possible values 1, 2, 3, 4, and randomly set starting value
    
    Methods:
     mydie.size()           # return the number of sides of mydie
     mydie.get_value()      # return the value of mydie 
     mydie.set_value(2)     # set the current value of mydie to 2
     mydie.roll()           # randomly set a new value of mydie and return that value
    """
    
    def __init__(self, num_sides):
        self.num_sides = num_sides
        self.roll()
        print(f"Die created with {self.num_sides} sides and value {self.value}")
        
    def roll(self):
        self.value = np.random.randint(1,self.num_sides+1)
        return self.value
        
    def size(self):
        return self.num_sides 
    
    def get_value(self):
        return self.value 
        
    def set_value(self, value):
        self.value = value 
        
```
