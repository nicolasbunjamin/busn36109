We model the spread of a pandemic with a simulation.   Our model will assume a fixed number of people for the length of the simulation.  Each day a person is in one of five states of health:

### States of our model

- (S) **S**usceptible to catching it when exposed.  Currently healthy, does not have, and never has had the disease.
- (C) **C**ontagious to others.  Currently has the disease but is aymptomatic.
- (H) Currently has the disease and knows it, has tested positive and/or is showing symptoms, and so is quarantined at **H**ome or **H**ospitalized and so is no longer considered a risk to pass the disease onto others.  We assume no capacity limits at the Hospital.
- (I) Has survived the disease and is now for this disease is considered **I**mmune and is assumed to remain so.  Those in this group are not quarantined and thus are active in the population.
- (D) **D**eceased

## Interpreting the States and Transitions

An individual would not be able to discern between being in state (S) or (C), and may have had the disease without symptoms, recovered, and be in state (I).  That is, we are studying how a disease will move through a population, even though a particular individual may not know what state they are in.  

So those in (C) will either be tested or exhibit symptoms and transition to (H), or unbeknowst to themselves will fight off the disease and transition directly to (I) where they are no longer contageous.  Those in (H) will either beat the disease and transition to (I) or succumb and transition to (D).

Each *day* we model that a person will transistion from each of these states to possibly some others or back to its current state.  The following transitions are possible:

From S, one can only remain in S or move to C
- `P_S_S` = Probability of remaining in S, healthy without contracting the disease 
- `P_S_C` = Probability of moving from S to C, contracting the disease

From C, one can only remain in C or move to H or I 
- `P_C_C` = Probability of remaining in C,  with the disease but is not aware or asymptomatic
- `P_C_H` = Probability of moving from C to H, moving from asymptomatic to aware and/or symptomatic 
- `P_C_I` = Probability of moving from C to I, fighting off the disease and becoming immune 

From H, one can only remain in H or move to D or I 
- `P_H_H` = Probability of remaining in H, continuing to test positive and/battling the disease 
- `P_H_D` = Probability of moving from H to D, succumbing to the disease 
- `P_H_I` = Probability of moving from H to I, successfully fighting off the disease and becoming immune

Once in I or D one remains there
- `P_I_I` = Probability of remaining in I is 1 
- `P_D_D` = Probability of remaining in D is 1

## Modeling the transition probabilities

The values of the transition probabilities model how much contact the population has with each other, whether precautions like masking or social distancing are being used, the characteristics of the disease (how contagious, how long one remains contagious), and the severity of the disease and viability of our health-care system (how many survive once symptomatic).

### Transitioning from Susceptible State S

We will consider the following definitions and factor to help define how the population moves from S to C.

- `prop_C` The proportion of the population that is not quarantined that is currently Contagious:

$$
\text{Prop}_\text{Contagious} \, = \, 
\frac{\text{\# Contagious}}{\text{\# Contagious + \# Susceptible + \# Immune}}
$$

-  `f` An exposure factor in (0,1), which models how contangious the disease is *and* how many interactions a person has with contagious people.  

We then define:

`P_S_C` =  `f` * `prop_C`

So as the proportion of the population that is contagious increases (`prop_C`), the probability of a Susceptible (S) person moving to Contagious (C) increases.  The `f` factor models the chance each day that such an encounter results in catching the disease and how many Contagious people a Susceptible person encounters in a day. 

So, a small value of `f` models that people are self-isolating, wearing masks, and/or that the disease is not very contagious.  Conversely, a large value of `f` models that people are interacting without precautions on a daily basis with many others and/or that the disease is very contagious.  The `f` factor would be studied and estimated for various populations (a dense city vs sparse countryside) and various diseases (airborne and highly contagious or requires intimate contact). 

Furthermore, a person who is in state S must either remain in S or transition to C, and so we must have:

`P_S_S` =  1 - (`f` * `Prop_C`)

```python
import numpy as np
from simpy_helpers import Entity, Resource, Source, Stats
import simpy
import matplotlib.pyplot as plt
```

```python
SIM_LENGTH = 200 

POPULATION_SIZE = 1000

NUMBER_INITIALLY_CONTAGEOUS = 40 

MIN_H_STAY_DURATION = 5

f = 0.20

P_C_C = 0.90
P_C_H = 0.05
P_C_I = 1.0 - P_C_C - P_C_H 

P_H_H = 0.42
P_H_I = 0.10
P_H_C = 0.42
P_H_D = 1.0 - P_H_H - P_H_I - P_H_C

P_D_D = 1 
P_I_I = 1 
```

```python
# Here is the Resource subclass
class Susceptible(Resource):
    def service_time(self, entity):
        return 1 
    
class Contagious(Resource):
    def service_time(self, entity):
        return 1 
    
class Hospital(Resource):    
    def service_time(self, entity):
        return 1

class Immune(Resource):
    def service_time(self, entity):
        return 1 
    
class Deceased(Resource):
    def service_time(self, entity):
        return 1 
```

```python
# Here is the Source subclass
class GenerateJobs(Source):
    def interarrival_time(self):
        return 0.0
    
    def build_entity(self):
        # print(f"count: {self.count}")
        attributes = {}
        if self.count <= NUMBER_INITIALLY_CONTAGEOUS:
            # print(f'Contageous Job {self.count} created at {self.env.now}')
            attributes['state'] = "C"
        else:
            # print(f'Susceptible Job {self.count} created at {self.env.now}')
            attributes['state'] = "S"
        
        return Job(env, attributes)
```

```python
## Here is the Entity Subclass
class Job(Entity):
    def process(self):
        h_stay_duration = 0
        
        while True: 
            # Put person in correct state for day
            # print(f"Time: {env.now}  Entity: {self.name} STATES:  Susceptible: {susceptible.count}  Contageous: {contagious.count}  Hospital {hospital.count}  Immune: {immune.count}  Deceased: {deceased.count}")
            if self.attributes['state'] == 'I': 
                #print('To IMMUNE')
                yield self.wait_for_resource(immune)
                yield self.process_at_resource(immune)
                self.release_resource(immune)
            elif self.attributes['state'] == 'D':    
                #print('To DECEASED')
                yield self.wait_for_resource(deceased)
                yield self.process_at_resource(deceased)
                self.release_resource(deceased)
            elif self.attributes['state'] == 'H':    
                #print('TO HOSPITAL')
                yield self.wait_for_resource(hospital)
                yield self.process_at_resource(hospital)
                self.release_resource(hospital)
            elif self.attributes['state'] == 'C':    
                # print(f'{self.name} To CONTAGEOUS')
                yield self.wait_for_resource(contagious)
                yield self.process_at_resource(contagious)
                self.release_resource(contagious)
            else:        
                # print(f'{self.name} To SUSCEPTIBLE')
                yield self.wait_for_resource(susceptible)
                yield self.process_at_resource(susceptible)
                self.release_resource(susceptible)
            
            # print("\n")
            # print(f"Time: {env.now}  COMPUTE NEXT STATE for Entity: {self.name} Susceptible: {susceptible.count}  Contageous: {contagious.count}  Hospital {hospital.count}  Immune: {immune.count}  Deceased: {deceased.count}")
            # Determine the person's next state 
            if self.attributes['state']=='C':
                self.attributes['state'] = np.random.choice(["C", "H", "I"], 
                                                            p=[P_C_C, P_C_H, P_C_I])
            elif self.attributes['state']=='H':
                
                # 1. Stay for the minimum duration (MIN_H_STAY_DURATION) 
                if h_stay_duration >= MIN_H_STAY_DURATION:

                    # 1a. After minimum stay duration, potentially get discharged
                    # NOTE: UNDER THIS LOGIC, CANNOT DIE BEFORE MIN_H_STAY_DURATION EITHER
                    self.attributes['state'] = np.random.choice(["H", "I", "C", "D"], 
                                                            p=[P_H_H, P_H_I, P_H_C, P_H_D])
                    
                    # 1b. Potentially stay for longer
                    if self.attributes['state'] == "H":
                        h_stay_duration += 1

                    # 1c. If discharged, reset hospital stay duration
                    else:
                        h_stay_duration = 0

                # 2. If not stayed for MIN_H_STAY_DURATION, increment hospital stay duration
                else:
                    h_stay_duration += 1
                    self.attributes['state'] = "H"
                    
            elif self.attributes['state']=='S':
                prop_c = contagious.count/(contagious.count + susceptible.count + immune.count)
                P_S_C = f * prop_c
                self.attributes['state'] = np.random.choice(["S", "C"], 
                                                            p=[1.0-P_S_C, P_S_C])
```

```python
np.random.seed(3298)
env = simpy.Environment()

immune = Immune(env, capacity=POPULATION_SIZE) 
contagious = Contagious(env, capacity=POPULATION_SIZE) 
susceptible = Susceptible(env, capacity=POPULATION_SIZE) 
deceased = Deceased(env, capacity=POPULATION_SIZE) 
hospital = Hospital(env, capacity=POPULATION_SIZE) 

source = GenerateJobs(env, number=POPULATION_SIZE) 

env.process(source.start(debug=False))
env.run(until=SIM_LENGTH)
```

```python
s = Stats.number_being_processed_over_time(susceptible)
i = Stats.number_being_processed_over_time(immune)
d = Stats.number_being_processed_over_time(deceased)
c = Stats.number_being_processed_over_time(contagious)

fig = plt.figure()
plt.plot(s, color="blue", label="susceptible")
plt.plot(c, color="red", label="contagious")
plt.plot(i, color="green", label="immune")
plt.plot(d, color="brown", label="deceased")
plt.xlabel('Day');
plt.ylabel('Number');
plt.legend();
plt.show()
```
