```python
ages = [31, 28, 24, 41, 24, 28]
sum = 0
for age in ages: 
    sum = sum + age
    print(f"age = {age}")
print(f"The sum of ages is {sum} and the average is {sum/len(ages)}")
```

```python
students = [ {'age': 31, 'name': "Bob"}, {'age': 28, 'name': "Anuj"}, {'age': 24, 'name': "Shandra"}, {'age': 41, 'name': "Cindi"}, {'age': 24, 'name': "Zara"}, {'age': 28, 'name': "Dirk"} ]

for student in students: 
    print(f"The student {student['name']} is {student['age']} years old")
```

```python
student_dict = {
    '2143': {'age': 31, 'name': 'Bob'}, 
    '4321': {'age': 28, 'name': 'Anuj'}, 
    '8788': {'age': 24, 'name': 'Shandra'}, 
    '6543': {'age': 41, 'name': 'Cindi'}, 
    '9878': {'age': 24, 'name': 'Zara'}, 
    '7689': {'age': 28, 'name': 'Dirk'}
}

student_id = '6543'
print(f"The age of the student with ID {student_id} is {student_dict[student_id]['age']}")
```

```python
students = ["Bob", "Anuj", "Shandra", "Cindi", "Zara", "Dirk"]
courses = ["Decision models", "Micro", "Application Development", "Accounting"]

for student in students: 
    print(f"Process student {student}")
    for course in courses: 
        print(f"Let's check if {student} took {course}")
```
