We need to choose a weekly stocking level `X` in anticipation of uncertain weekly demand.

For each unit short for the week a cost of `C_s=20` is incurred.  For each unit stocked in excess of demand a cost of `C_e=10` is incurred.  

We have collected historical weekly demand data with the following equally likely 10 scenarios:


| Scenario  | weekly demand | 
|--|:--:|
| **0** | 230 | 
| **1** | 210 |
| **2** | 160 |
| **3** | 450 |
| **4** | 310 |
| **5** | 190 |
| **6** | 130 |
| **7** | 340 |
| **8** | 370 |
| **9** | 290 |

What stocking decision `X` should be made to minimize the expected cost?

```python
# Out data and cost parameters
demand = [ 230, 210, 160, 450, 310, 190, 130, 340, 370, 290 ]
C_s = 20
C_e = 10
```

```python
mean_demand = sum(demand)/len(demand)
print(f"Average demand: {mean_demand}")
```

We assume the optimal stocking decision X is between the min and max value of the demand scenario.

```python
x_min = min(demand)
x_max = max(demand)
print(f"We assume our stocking decision X is betweeen {x_min} and {x_max}")
```

**Example:  Suppose X=400, estimate the total cost, shortage plus excess**

```python
import random
```

```python
random.choice(demand) # Randomly choose an element from demand
```

```python
x = 400
n = 100 # Generate 100 random demand scenarios
total_cost = 0
for i in range(n):
    d = random.choice(demand)
    num_short = max(0, d - x)
    num_excess = max(0, x - d)
    cost = C_s * num_short + C_e * num_excess
    total_cost += cost
    # print(f"X: {x}     Demand: {d}     Shortage: {num_short}     Excess: {num_excess}     Cost: {cost}")
print(f"X: {x}     Average Cost: {total_cost/n}")
```

```python
n = 10**5
for x in range(x_min, x_max+10, 10):
    total_cost = 0
    for i in range(n):
        d = random.choice(demand)
        num_short = max(0, d-x)
        num_excess = max(0, x-d)
        cost = C_s * num_short + C_e * num_excess
        total_cost += cost
        # print(f"X: {x}     Demand: {d}     Shortage: {num_short}     Excess: {num_excess}     Cost: {cost}")
    print(f"X: {x}     Average Cost: {round(total_cost/n, 1)}")
```
