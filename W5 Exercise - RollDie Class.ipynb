### Create a `RollDie` Python Class


Usage Ex.  `mydie = RollDie(4)`   # create a 4 sided die with possible values 1,2,3,4, and randomly set starting value

The class should provide the following methods for an instance `mydie`:
 - `mydie.size()`: return the number of sides of mydie
 - `mydie.get_value()`:  return the current value of mydie
 - `mydie.set_value(2)`:  set the current value of mydie to 2
 - `mydie.roll()`:  randomly set a new value of mydie and return that value
 - `mydie.number_of_rolls()`:  return the total number of rolls for mydie 
 
The class should provide the following class methods:

 - `Rolldie.all_number_of_rolls()`:  return the total number of rolls for all die in class

```python
import numpy as np
class RollDie:
    
    """
    A class to roll and set a die of any number of sides.
    
    Usage:  Initialize an object, pass nunber of sides of die
    Ex.  mydie = RollDie(4)   # create an 4 sided die with possible values 1, 2, 3, 4, and randomly set starting value
    
    Class Methods:
    
    all_number_of_rolls()   # Number of rolls by entire class
    
    Instance Methods:
     mydie.size()           # return the number of sides of mydie
     mydie.get_value()      # return the value of mydie 
     mydie.set_value(2)     # set the current value of mydie to 2
     mydie.roll()           # randomly set a new value of mydie and return that value
     mydie.number_of_rolls()    # how many rolls by mydie 
    """

    all_num_rolls = 0
    all_instances = []
    
    def __init__(self, num_sides):
        self.num_sides = num_sides
        self.num_rolls = 0
        self.value = np.random.randint(1, num_sides + 1)
        print(f'Die created with {self.num_sides} sides. Value: {self.value}.')

    def size(self):
        return self.num_sides

    def roll(self):
        self.value = np.random.randint(1, self.size()+1)
        self.num_rolls += 1
        RollDie.all_num_rolls += 1
        return self.value

    def get_value(self):
        return self.value

    def set_value(self, value):
        if 1 <= value <= self.size():  # Works for both int and float
            self.value = value
            print(f'Die value set to {self.value}.')
        else:
            print(f'Die only has {self.num_sides} sides. Cannot set it to {value}')

    def number_of_rolls(self):
        return self.num_rolls

    @classmethod
    def all_number_of_rolls(cls):
        return cls.all_num_rolls
    
    @classmethod
    def reset_number_of_rolls(cls):
        cls.all_num_rolls = 0
        for die in cls.all_instances:
            die.num_rolls = 0
```
